```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = [
["S10000", 0.355769,  0.503401],
["S30000", 0.695946,  0.70068],
["S50000", 0.74359,   0.789116],
["S100000", 0.810127,  0.870748],
["PCM", 0.738693,  1],
["CMG200", 0.864706,  1],
["CMG100", 0.890909,  1],
["CMG010", 0.967105,  1]]

mem = [
["S10000",  2568368],
["S30000",  7686320],
["S50000",  12806320],
["S100000",  25605040],
["PCM",  1578045244],
["CMG200",  4252184],
["CMG100",  8274296],
["CMG010",  58664784]]
```

```python
df = pd.DataFrame(data, columns = ["Sketch", "Precision", "Recall"])
df
```

```python
memdf = pd.DataFrame(mem, columns = ["Sketch", "Memory Usage"])
memdf
```

```python
sns.set(style="whitegrid")
sns.set(font_scale=1.5)
```

```python
g = sns.barplot(x = "Sketch", y = "Precision", data=df)
g.set_xlabel("")

plt.xticks(rotation = 45)

plt.tight_layout()

plt.savefig('HH_precision.pdf')
```

```python
g = sns.barplot(x = "Sketch", y = "Recall", data=df)
g.set_xlabel("")

plt.xticks(rotation = 45)

plt.tight_layout()

plt.savefig('HH_recall.pdf')
```

```python
g = sns.barplot(x = "Sketch", y = "Memory Usage", data=memdf)
g.set_yscale("log")
g.set_xlabel("")
g.set_ylabel("Memory Usage (Byte)")

plt.xticks(rotation = 45)
plt.tight_layout()

plt.savefig('HH_mem.pdf')
```
